```python
import cv2
from IPython.display import display, Image

def load_and_display_image(image_path):
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    display(Image(data=cv2.imencode('.png', img_rgb)[1]))

load_and_display_image(load_and_display_image(r"C:\Users\User\Downloads\pic.jpg"))
```

```python
def crop_image(image_path, x, y, width, height):
    img = cv2.imread(image_path)
    cropped_img = img[y:y+height, x:x+width]

    cropped_img_rgb = cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB)

    display(Image(data=cv2.imencode('.png', cv2.cvtColor(img, cv2.COLOR_BGR2RGB))[1]))
    display(Image(data=cv2.imencode('.png', cropped_img_rgb)[1]))

crop_image(r"C:\Users\User\Downloads\pic.jpg", x=100, y=50, width=250, height=150)
```
